```python
marker_cluster = MarkerCluster()
```

```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Merging the class information into the launch_sites_df dataframe
launch_sites_df = launch_sites_df.merge(spacex_df.groupby('Launch Site')['class'].mean().reset_index(), on='Launch Site')

# Defining marker colors based on the success rate (class value)
# If the majority of launches (class >= 0.5) are successful, marker color is green; otherwise, red
launch_sites_df['marker_color'] = launch_sites_df['class'].apply(lambda x: 'green' if x >= 0.5 else 'red')

launch_sites_df

```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python

# Railway marker coordinates
railway_marker = [28.55752, -80.80155]

# Launch site coordinates
launch_coordinate = [28.57337, -80.64669]

# Calculating the distance
distance_coastline = calculate_distance(launch_coordinate[0], launch_coordinate[1], railway_marker[0], railway_marker[1])
distance_coastline
```

```python
# Creating a new map for clarity
distance_map = folium.Map(location=[(railway_marker[0] + launch_coordinate[0]) / 2, 
                                    (railway_marker[1] + launch_coordinate[1]) / 2], zoom_start=10)

# Adding a marker for the railway point
folium.Marker(
    railway_marker,
    icon=folium.Icon(color='blue', icon='info-sign'),
    popup='Railway Marker'
).add_to(distance_map)

# Adding a marker for the launch site
folium.Marker(
    launch_coordinate,
    icon=folium.Icon(color='green', icon='info-sign'),
    popup='Launch Site'
).add_to(distance_map)

# Adding the distance marker
distance_marker = folium.Marker(
    railway_marker,
    icon=folium.DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12; color:#d35400;"><b>{distance_coastline:.2f} KM</b></div>'
    )
)
distance_marker.add_to(distance_map)

distance_map

```

*TODO:* Draw a `PolyLine` between a launch site to the selected coastline point


```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
# lines=folium.PolyLine(locations=coordinates, weight=1)

# Adding a PolyLine between the launch site and the coastline point
folium.PolyLine(
    locations=[launch_coordinate, railway_marker],
    color='red',
    weight=2,
    opacity=1
).add_to(distance_map)

distance_map

```
